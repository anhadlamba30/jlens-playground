# Download Jacobian Lens files


```python
from huggingface_hub import HfApi, snapshot_download
REPO_ID='neuronpedia/jacobian-lens'
files=HfApi().list_repo_files(REPO_ID, repo_type='model')
folders=sorted({f.split('/')[0] for f in files if '/' in f and not f.startswith('.')})
folders
```

```python
MODEL_FOLDER='llama3.1-8b'
snapshot_download(repo_id=REPO_ID, repo_type='model', allow_patterns=[f'{MODEL_FOLDER}/**'], local_dir='../lenses')
```
